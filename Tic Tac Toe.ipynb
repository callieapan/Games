```python
from random import *
import sys
print "starting Tic Tac Toe"
print "\n"
playerA = raw_input('enter player A name:').strip()
playerB = raw_input('enter player B name:').strip()
currentPlayer = 0 # current player is 0 at first, it will be either 1 or 2 
playerNames = [playerA, playerB]
Board = []
MovesMade = 0
Marks = ['A', 'B']
GameOn = True
Columns=[[0,0,0],[0,0,0],[0,0,0]]
Horizontals=[[],[]]



def MakeBoard():
    global Board
    #0 elements on the board, 0 means empty 1 means player 1 in it, 2 means player 2 in it
    print "making board for Tic Tac Toe..."
    Board = [[1,2,3],[4,5,6],[7,8,9]]
        
        
def PrintBoard():
    global Board
    print "\n"
    for i in Board:
        print i

def randFirstPlayer():
    global currentPlayer
    n = random()
    if n>0.5:
        currentPlayer = 1
    else: 
        currentPlayer = 2
    
        

def makeMove():
    global Board
    global MovesMade
    
    #check who is current player, ask for input
    while True: 
        print "Player {}'s turn, please input the square number (1 to 9) where you want to place your mark".format(playerNames[currentPlayer-1])
        move = int(raw_input().strip())
        if not (1<= move <= 9):
            print "wrong entry please enter 1 through 9"
            continue
        a,b= convert(move)
        if Board[a][b] == 'A' or Board[a][b] == 'B':
            print "wrong entry, {} is already occupied by {}".format (str(move), Board[move])
            continue 
        else: break
        #if none of the above situations then make the new mark
    Board[a][b] = Marks[currentPlayer-1]
    MovesMade +=1
    

def convert(move):
    #change 1 to 9 into coordinates
    if move%3 == 0: #if move is 3, 6, or 9
        a = move/3-1
        b = 2
    else:
        a = move/3
        b = move%3 -1
        
    return a,b   

    
def checkWin():
    global Board
    global Columns
    global Horizontals
    mark = Marks[currentPlayer-1]
    #print "current mark ", mark
    #check rows
    for i in Board:
        #print "check row", i
        if i.count(mark) == 3:
            GameEnd() 
            break
    #check columns
    #make columns first, then cheeck
    x = 0
    for i in Board:
        Columns[0][x] = i[0]
        Columns[1][x] = i[1]
        Columns[2][x]= i[2]
        x+=1
    for i in Columns:
        #print 'check columns', i
        if i.count(mark)==3:
            GameEnd()
            break
            
    #check horiz
    Horizontals[0] = [Board[0][0],Board[1][1],Board[2][2]]
    Horizontals [1]=[Board[0][2], Board[1][1], Board[2][0]]
    for i in Horizontals:
        #print 'check horizontal', i
        if i.count(mark)==3:
            GameEnd()
            break
            
    if MovesMade == 9:
        GameTied()
    
def switchPlayer():
    global currentPlayer    
    if currentPlayer ==1:
        currentPlayer =2
    else: currentPlayer = 1

def GameEnd():
    global GameOn
    # print the outcome
    print "Game End! {} won!".format(playerNames[currentPlayer-1])
    GameOn = False
    sys.exit()

def GameTired():
    global GameOn
    print "Game Tied"
    GameOn = False
    sys.exit()

MakeBoard()
PrintBoard()
randFirstPlayer()

while GameOn:
    makeMove()
    PrintBoard()
    checkWin()
    switchPlayer()
        
    
```
